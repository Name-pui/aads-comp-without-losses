```python
import os
import struct
from PIL import Image

def convert_to_custom_raw(input_path, output_path):
    img = Image.open(input_path)
    
    if img.mode in ('RGBA', 'P'):
        img = img.convert('RGB')
    
    width, height = img.size
    channels = 3 if img.mode == 'RGB' else 1
    
    header = struct.pack('<IIB', width, height, channels)
    header += b'\x00' * 3  # Добиваем до 12 байт для ровности

    with open(output_path, 'wb') as f:
        f.write(header)
        f.write(img.tobytes())
        
    return os.path.getsize(output_path)

def analyze(original_path, raw_path):
    size_orig = os.path.getsize(original_path)
    size_raw = os.path.getsize(raw_path)
    ratio = size_raw / size_orig

    print(f"Исходный файл: {original_path} ({size_orig} байт)")
    print(f"RAW файл:      {raw_path} ({size_raw} байт)")
    print(f"Коэффициент сжатия:  {ratio:.2f}")


input_file = 'test4.jpg' 
output_file = 'test4.raw'

convert_to_custom_raw(input_file, output_file)
analyze(input_file, output_file)
print("\n")

input_file = 'test5.jpg' 
output_file = 'test5.raw'

convert_to_custom_raw(input_file, output_file)
analyze(input_file, output_file)
print("\n")

input_file = 'test6.jpg' 
output_file = 'test6.raw'

convert_to_custom_raw(input_file, output_file)
analyze(input_file, output_file)
```
